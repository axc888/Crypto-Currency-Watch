```python
import matplotlib.pyplot as plt
from Tkinter import *
import requests
import json
import os
os.system('cls')
root = Tk()

header_name = Label(root, text="Name", bg="cyan", font = "Verdana 8 bold")
header_name.grid(row=0, column=0, sticky=N+S+E+W)

header_rank = Label(root, text="Rank", bg="cyan", font = "Verdana 8 bold")
header_rank.grid(row=0, column=1, sticky=N+S+E+W)

header_current_price = Label(root, text="Current Price", bg="cyan", font = "Verdana 8 bold")
header_current_price.grid(row=0, column=2, sticky=N+S+E+W)

header_price_paid = Label(root, text="Price BTC", bg="cyan", font = "Verdana 8 bold")
header_price_paid.grid(row=0, column=3, sticky=N+S+E+W)

header_1_hr_change = Label(root, text="1 HR Change", bg="cyan", font = "Verdana 8 bold")
header_1_hr_change.grid(row=0, column=5, sticky=N+S+E+W)

header_24_hr_change = Label(root, text="24 HR Change", bg="cyan", font = "Verdana 8 bold")
header_24_hr_change.grid(row=0, column=6, sticky=N+S+E+W)

header_7_day_change = Label(root, text="7 Day Change", bg="cyan", font = "Verdana 8 bold")
header_7_day_change.grid(row=0, column=7, sticky=N+S+E+W)

api_request = requests.get("https://api.coinmarketcap.com/v1/ticker/")
api = json.loads(api_request.content)


def show_colour(amount):
    if amount >= 0:
        return "green"
    else:
        return "red"

def update_crypto():
    api_request = requests.get("https://api.coinmarketcap.com/v1/ticker/")
    api = json.loads(api_request.content)
    r = requests.get('https://api.coinmarketcap.com/v1/ticker/')
    row_count = 1
    pie = []
    pie_size = []
    for x in r.json():
 
        name = Label(root, text=x["name"], bg="white")
        name.grid(row=row_count, column=0, sticky=N+S+E+W)

        rank = Label(root, text=x["rank"], bg="white")
        rank.grid(row=row_count, column=1, sticky=N+S+E+W)

        current_price = Label(root, text="${0:.2f}".format(float(x["price_usd"])), bg="white")
        current_price.grid(row=row_count, column=2, sticky=N+S+E+W)

        market_cap_usd = Label(root, text="${0:.6f}".format(float(x["price_btc"])), bg="white")
        market_cap_usd.grid(row=row_count, column=3, sticky=N+S+E+W)


        one_hr_change = Label(root, text="${0:.2f}".format(float(x["percent_change_1h"])), bg="white", fg=show_colour(float(x["percent_change_1h"])))
        one_hr_change.grid(row=row_count, column=5, sticky=N+S+E+W)

        tf_hr_change = Label(root, text="${0:.2f}".format(float(x["percent_change_24h"])), bg="white", fg=show_colour(float(x["percent_change_24h"])))
        tf_hr_change.grid(row=row_count, column=6, sticky=N+S+E+W)

        seven_day_change= Label(root, text="${0:.2f}".format(float(x["percent_change_7d"])), bg="white", fg=show_colour(float(x["percent_change_7d"])))
        seven_day_change.grid(row=row_count, column=7, sticky=N+S+E+W)
        
        pie.append(x["name"])
        pie_size.append(x["price_btc"])

        row_count += 1 
        if row_count==26:
            break;

    root.title("Crypto Currency Watch")

update_button = Button(root, text="Update Prices", command=update_crypto)
update_button.grid(row=row_count, column=7, sticky=E+S, padx=10, pady=10)
update_crypto()
root.mainloop()    
```
